```python
import pandas as pd
import numpy as np
from pycaret.classification import *
import warnings
warnings.filterwarnings('ignore')

print("🚀 PYCARET SETUP COMPLETO - FULL POWER")
print("=" * 60)

# ===== CARREGAR DADOS =====
train_path = '/Users/marcelosilva/Desktop/projectOne/5/A-TrainTestDataset/TrainDS.csv'
df_train = pd.read_csv(train_path)
print(f"✅ Dataset: {df_train.shape}")

# ===== VERIFICAR MEMÓRIA DISPONÍVEL =====
import psutil
memory_gb = psutil.virtual_memory().available / (1024**3)
print(f"💾 Memória disponível: {memory_gb:.1f} GB")

# ===== SETUP COMPLETO E ROBUSTO =====
print(f"\n🔧 CONFIGURANDO SETUP COMPLETO (32GB RAM)")
print("=" * 60)

try:
    clf = setup(
        data=df_train,
        target='status_nutricional_who',
        
        # 🚫 Features a ignorar
        ignore_features=['id_anon', 'vd_zimc'],
        
        # 📊 VARIÁVEIS ORDINAIS (CRÍTICO!)
        ordinal_features={
            'def_idade_gest': [0, 1, 2],           # prematuro < adequado < pos_termo
            'adequacao_prenatal': [0, 1, 2],       # ausente < insuficiente < adequado
            'idade_mae_cat': [0, 1, 2],            # jovem < adulta < madura
            'peso_cat': [0, 1, 2],                 # baixo < normal < alto
            'classificacao_peso': [0, 1, 2]        # PIG < AIG < GIG
        },
        
        # 🎯 BALANCEAMENTO (CLASSE MINORITÁRIA 2.4%)
        fix_imbalance=True,
        fix_imbalance_method='smote',
        
        # 🔧 PREPROCESSING COMPLETO
        remove_multicollinearity=True,
        multicollinearity_threshold=0.95,
        feature_selection=True,
        feature_selection_method='univariate',
        transformation=True,
        normalize=True,
        
        # 📊 VALIDAÇÃO ROBUSTA
        fold=10,                               # 10-fold CV (robusto!)
        fold_strategy='stratifiedkfold',       # Manter proporções
        
        # ⚡ PERFORMANCE
        train_size=0.8,                        # 80% treino, 20% validação
        session_id=123,                        # Reprodutibilidade
        use_gpu=False,                         # CPU (M1 é potente)
        n_jobs=-1,                            # Usar todos os cores
        
        # 📈 CONFIGURAÇÕES AVANÇADAS  
        profile=False,                         # Economizar tempo
    )
    
    print("✅ SETUP COMPLETO FUNCIONOU!")
    
    # Verificar configuração
    print(f"\n📊 INFORMAÇÕES DO SETUP:")
    print(f"   Shape transformado: {get_config('X_train').shape}")
    print(f"   SMOTE aplicado: {'Sim' if get_config('fix_imbalance') else 'Não'}")
    print(f"   Folds: {get_config('fold')}")
    
except Exception as e:
    print(f"❌ ERRO no setup completo: {e}")
    exit()
# ===== VER FEATURES SELECIONADAS ===== 👈 COLE AQUI!
print("🔍 FEATURES QUE PASSARAM PELA SELEÇÃO:")
print("=" * 60)

# Features que sobreviveram
selected_features = get_config('X_train').columns
print(f"✅ FEATURES SELECIONADAS ({len(selected_features)}):")
for i, feature in enumerate(selected_features, 1):
    print(f"   {i:2d}. {feature}")

# Features originais (sem target e ignoradas)
original_features = df_train.drop(['status_nutricional_who', 'id_anon', 'vd_zimc'], axis=1).columns
print(f"\n📊 ESTATÍSTICAS:")
print(f"   Features originais: {len(original_features)}")
print(f"   Features selecionadas: {len(selected_features)}")
print(f"   Taxa de retenção: {(len(selected_features)/len(original_features))*100:.1f}%")

# Features que foram removidas
removed_features = set(original_features) - set(selected_features)
print(f"\n❌ FEATURES REMOVIDAS ({len(removed_features)}):")
for i, feature in enumerate(sorted(removed_features), 1):
    print(f"   {i:2d}. {feature}")

print("\n🎯 ANÁLISE DAS FEATURES SOBREVIVENTES:")
# ===== FIM DO CÓDIGO NOVO =====    

# ===== COMPARAÇÃO COMPLETA DE MODELOS =====
print(f"\n🤖 COMPARAÇÃO COMPLETA - TODOS OS MODELOS")
print("=" * 60)

# TODOS OS MODELOS FUNCIONANDO (16 modelos!)
all_models = [
    # Tree-based (melhores para dados estruturados)
    'rf',           # Random Forest
    'et',           # Extra Trees  
    'dt',           # Decision Tree
    'xgboost',      # XGBoost
    'gbc',          # Gradient Boosting
    'ada',          # AdaBoost
    
    # Linear models
    'lr',           # Logistic Regression
    'ridge',        # Ridge Classifier
    'lda',          # Linear Discriminant Analysis
    'qda',          # Quadratic Discriminant Analysis
    
    # SVM
    'svm',          # SVM Linear
    'rbfsvm',       # SVM RBF
    
    # Instance-based
    'knn',          # K-Nearest Neighbors
    
    # Probabilistic
    'nb',           # Naive Bayes
    'gpc',          # Gaussian Process
    
    # 🧠 Deep Learning
    'mlp'           # Neural Network
]

print(f"🎯 Testando {len(all_models)} modelos com setup completo...")
print("📋 Modelos incluídos:")
for i, model in enumerate(all_models, 1):
    print(f"   {i:2d}. {model}")

print(f"\n⏳ Executando comparação completa (pode demorar 5-10 minutos)...")
print("🎯 Configuração: 10-fold CV + SMOTE + variáveis ordinais + feature selection")

try:
    # 🚀 COMPARAÇÃO COMPLETA COM TODAS CONFIGURAÇÕES!
    comparison_full = compare_models(
        include=all_models,
        sort='F1',                    # F1-score para dados desbalanceados
        n_select=10,                  # Top 10 modelos
        fold=10,                      # 10-fold CV (robusto!)
        verbose=True,                 # Mostrar progresso
        turbo=False                   # Não economizar - usar configuração completa
    )
    
    print("\n🏆 COMPARAÇÃO COMPLETA FINALIZADA!")
    print("📊 Resultados com TODAS as configurações aplicadas:")
    print("   ✅ 10-fold Cross Validation")
    print("   ✅ SMOTE para classe minoritária") 
    print("   ✅ Variáveis ordinais preservadas")
    print("   ✅ Feature selection aplicada")
    print("   ✅ Multicolinearidade removida")
    print("   ✅ Normalização aplicada")
    
    print(f"\n🥇 MELHOR MODELO: {type(comparison_full[0]).__name__}")
    
except Exception as e:
    print(f"❌ ERRO na comparação completa: {e}")
    print("🔧 Tentando versão reduzida...")
    
    # Fallback: modelos mais estáveis
    stable_models = ['rf', 'xgboost', 'gbc', 'lr', 'nb', 'mlp', 'et', 'ada']
    
    try:
        comparison_stable = compare_models(
            include=stable_models,
            sort='F1',
            n_select=8,
            fold=10,
            verbose=True
        )
        print("✅ COMPARAÇÃO ESTÁVEL FUNCIONOU!")
        
    except Exception as e2:
        print(f"❌ ERRO também na versão estável: {e2}")

print(f"\n🎯 COMPARAÇÃO COMPLETA FINALIZADA!")
print("📈 Agora os resultados devem ser MUITO melhores!")
```
